```python
import tensorflow as tf
```

In this notebook I will use autodif feature of tensor flow to detemine the slope of $y=x^2$ when $x = 5$.

```python
x = tf.Variable(5.0) #define the value of the variable that you want to wathch
```

```python
with tf.GradientTape() as t:
  t.watch(x)
  y = x**2
t.gradient(y,x).numpy() #autodifferentiates and computes the derivative at point x
```

```python
import torch
```

In this notebook I will deffirinate equation


>$
y = x^2
$

and find the slop of this equation when $x=5 using 'automatic differentiation' in pytorch. That is I am not going to expicitly derive the first derivative, but contigously track the gradient of x in the equation.






```python
x = torch.tensor(5.0)
x.requires_grad_() #tells it to track gradient
y = x**2 #define the equation
y.backward() # autodifferentiates y
x.grad # provides the gradient (slop) at point x
```

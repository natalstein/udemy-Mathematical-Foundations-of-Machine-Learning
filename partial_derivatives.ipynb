```python
import numpy as np
import matplotlib.pyplot as plt
import torch
import math
```

Define $f(x,y)$ for $z= x^2 -y^2$

```python
def f(my_x, my_y):
  return my_x**2 + my_y**2
```

plot z with respect to x with y as a constant

```python
xs = np.linspace(-3,3,1001)
y = 0
plt.plot(xs,f(xs,y))
plt.show()
```

Determine the slope of z at a given points x we first derive and define the partial derivative of z with respect to x. $$\frac{\delta z}{\delta x} = 2x$$

```python
def delz_delx(my_x, my_y):
  return 2*my_x
```

```python
x_given = [-2, -1, 0, 1, 2]
colors = ['red', 'orange', 'green', 'blue', 'purple']
```

```python
def point_and_tangent_wrt_x(my_xs, my_x, my_y, my_f, fprime, col):
  my_z = my_f(my_x, my_y)
  plt.scatter(my_x, my_z, color=col, zorder = 3)

  tangent_m = fprime(my_x, my_y)
  tangent_b = my_z - tangent_m*my_x
  tangent_line = tangent_m*my_xs + tangent_b
  plt.plot(my_xs, tangent_line, color=col, linestyle = 'dashed', linewidth =0.7, zorder = 3)
```

```python
fig, ax = plt.subplots(figsize=(5,3))
plt.axvline(x=0, color='black')
plt.axhline(y=0, color='black')
plt.plot(xs,f(xs,y))
for i, x in enumerate(x_given):
  point_and_tangent_wrt_x(xs, x, y, f, delz_delx, colors[i])
plt.ylim(-1,9)
```

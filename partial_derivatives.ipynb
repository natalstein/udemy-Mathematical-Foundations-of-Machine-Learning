```python
import numpy as np
import matplotlib.pyplot as plt
import torch
import math
```

Define $f(x,y)$ for $z= x^2 -y^2$

```python
def f(my_x, my_y):
  return my_x**2 - my_y**2
```

plot z with respect to x with y as a constant

```python
xs = np.linspace(-3,3,1001)
y = 0
plt.plot(xs,f(xs,y))
plt.show()
```

Determine the slope of z at a given points x we first derive and define the partial derivative of z with respect to x. $$\frac{\delta z}{\delta x} = 2x$$

```python
def delz_delx(my_x, my_y):
  return 2*my_x
```

```python
x_given = [-2, -1, 0, 1, 2]
colors = ['red', 'orange', 'green', 'blue', 'purple']
```

```python
def point_and_tangent_wrt_x(my_xs, my_x, my_y, my_f, fprime, col):
  my_z = my_f(my_x, my_y)
  plt.scatter(my_x, my_z, color=col, zorder = 3)

  tangent_m = fprime(my_x, my_y)
  tangent_b = my_z - tangent_m*my_x
  tangent_line = tangent_m*my_xs + tangent_b
  plt.plot(my_xs, tangent_line, color=col, linestyle = 'dashed', linewidth =0.7, zorder = 3)
```

```python
fig, ax = plt.subplots(figsize=(7,3))
plt.axvline(x=0, color='black')
plt.axhline(y=0, color='black')
plt.plot(xs,f(xs,y))
for i, x in enumerate(x_given):
  point_and_tangent_wrt_x(xs, x, y, f, delz_delx, colors[i])
plt.ylim(-1,9)
```

Now, I will to the same for z with respect to y
$$\frac{\delta z}{\delta y} = 2y$$

```python

ys = np.linspace(-3,3,1001)
x = 0
def delz_dely(my_y, my_x):
  return -2*my_y
y_given = [-2, -1, 0, 1, 2]

def point_and_tangent_wrt_x(my_ys, my_y_given, my_x,my_f, my_fprime):
  plt.figure(figsize=(3,3))
  plt.xlim(np.min(y_given)-1,np.max(y_given)+1)
  plt.ylim(-6,np.max(f(my_x, my_ys)+1))
  plt.axvline(x=0, color='black', linewidth = 1, alpha = 0.5)
  plt.axhline(y=0, color='black', linewidth = 1, alpha =0.5)
  for i, y in enumerate(my_y_given):
    plt.scatter(y,f(my_x, y))
    m = my_fprime(y, my_x)
    b = my_f(my_x, y) - m*y
    line = m*my_ys + b
    plt.plot(ys, line, linestyle = 'dashed')
  plt.plot(my_ys,my_f(x,my_ys))
  plt.show()
```

```python
point_and_tangent_wrt_x(ys, y_given, x,f, delz_dely)
```
